## **Question 1 Which movies have maximum views/ratings?**


> Explanation

1. First I drop the user_id column which was not relevant for us in this task.

2. Then I count the number of the ratings over the Movie columns & named it as `rate_count`.
3. Sort the dataframe in the descending order to get the maximum rating value on the Top. 




```
df = df.drop(['user_id'],axis=1)
df.head()
```

```
movie_rate = pd.DataFrame()
movie_rate['rate_count'] = df.count(axis=0)
movie_rate.head()
```

```
movie_rating = movie_rate.sort_values("rate_count",ascending = False)
movie_rating.head(1)
```

# **Question 2 What is the average rating for each movie? Define the top 5 movies with the maximum ratings.**


> Explaination


1.   First I created a new column which consists of average value of the rating while skipping the NaN values.
Named this new column as `avg_movie_rating`
2.   Now, Sort the table dataframe according to the number of ratings 




```
avg_movie_rating = df.mean(axis = 0, skipna = True) 
```

```
movie_rate['avg_movie_rating'] = avg_movie_rating
movie_rate
```

```
sort_avg_movie_rating = movie_rate.sort_values(["rate_count"],ascending = False)
sort_avg_movie_rating.head(5)
```

# **Question 3 Define the top 5 movies with the least audience.**

```
sort_avg_movie_rating = movie_rate.sort_values(["rate_count"],ascending = True)

sort_avg_movie_rating.head(5)
```

```
df = pd.read_csv('Amazon - Movies and TV Ratings.csv')
df = df.fillna(value=0)
df
```

```
df_new = pd.melt(df,id_vars=["user_id"])
df_new.head()
```

```
df_new.shape
```

```
df_new = df_new.sort_values(["user_id"],ascending = True)
df_new
```

```
!pip install scikit-surprise
```

```
from surprise import Reader, Dataset
reader = Reader()
data = Dataset.load_from_df(df_new[[ 'user_id','variable', 'value']], reader)
```

```
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=0.25)
```

# **Question 5 Build a recommendation model on training data**

```
from surprise import SVD, accuracy
algo = SVD()
algo.fit(trainset)
```

# **Question 6 Make predictions on the test data**

```
predictions = algo.test(testset)
```

```
accuracy.rmse(predictions)
```

```
for i in range(0,10):
    print("User " + predictions[i][0]+ " should be recommended "+ predictions[i][1])
    print()
    
```
